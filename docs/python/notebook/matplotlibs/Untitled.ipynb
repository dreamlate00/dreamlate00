```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.title("title")#括号当中输入标题的名称
plt.show()
```

```python
plt.rcParams['font.sans-serif']=['SimHei']
plt.title("标题")
plt.show()
```

```python
plt.figure(figsize=(6, 3))
plt.plot(6, 3)
plt.plot(3, 3 * 2)
plt.show()
```

```python
# 坐标轴
plt.title("坐标轴")
plt.xlim(0,6) #x轴坐标轴
plt.ylim((0, 3))#y轴坐标轴
plt.xlabel('X')#x轴标签
plt.ylabel('Y')#y轴标签
plt.show()
```

```python
# 负数的坐标轴
plt.title("坐标轴")
plt.rcParams['axes.unicode_minus']=False
plt.xlim(-6,6) #x轴坐标轴
plt.ylim((-3, 3))#y轴坐标轴
plt.xlabel('X')#x轴标签
plt.ylabel('Y')#y轴标签
plt.show()
```

```python
# 设置 label 和 legend
plt.plot(2, 3, label="123")#第一个label
plt.plot(2, 3* 2, label="456")#第二个label
plt.legend(loc='best')#图列位置，可选best，center等
plt.show()
```

```python
# 添加注释
x=np.linspace(0,10,200)#从0到10之间等距产生200个值
y=np.sin(x)

plt.plot(x,y,linestyle=':',color='b')
plt.annotate(text='标记点',xy=(3,np.sin(3)),xytext=(4,-0.5),weight='bold',color='b',arrowprops=dict(arrowstyle='-|>',color='k'))
plt.show()
```

### 子图

```python
ax1 = plt.subplot(2, 2, 1)
plt.plot(x,np.sin(x), 'k')

ax2 = plt.subplot(2, 2, 2, sharey=ax1) # 与 ax1 共享y轴
plt.plot(x, np.cos(x), 'g')

ax3 = plt.subplot(2, 2, 3)
plt.plot(x,x, 'r')

ax4 = plt.subplot(2, 2, 4, sharey=ax3) # 与 ax3 共享y轴
plt.plot(x, 2*x, 'y')
```

### 散点图

```python
x = np.random.rand(10)
y = np.random.rand(10)
plt.scatter(x,y)
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
s = (30 * np.random.rand(50))**2
# s = 60
plt.scatter(x, y,s, c=colors, alpha=0.5)
plt.show()
```
